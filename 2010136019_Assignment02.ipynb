## 3. 키보드로 정수값을 입력 받고 그 값이 양수인지 음수인지를 비트 연산자를 이용하여 판단하는 프로그램을 작성하시오. 

```python
value = input("Enter integer: ")

if value & 0x80 == 0x80:
    print("음수입니다.")
elif value & 0x80 == 0x00:
    print("양수입니다.")
```

## 4. 두 개의 리스트를 인자로 받아서 그 두 개의 리스트에 대한 '합집합'을 반환하는 함수 list_union(lista, listb)를 작성하시오.

```python
def list_union(lista, listb):
    n1 = len(lista) # lista의 길이를 반환
    n2 = len(listb) # listb의 길이를 반환
    i=0
    
    if n1>n2:  #탐색을 효율적으로 하기 위해 길이가 짧은 리스트를 중복탐색 대상으로 지정한다.
        while i<n2:
            if (listb[i] in lista) == False:
                lista.append(listb[i]) 
            i+=1
        
        lista.sort()
        return lista
    else :    
        while i<n1:
            if (lista[i] in listb) == False:
                listb.append(lista[i]) 
            i+=1
    
    listb.sort()
    return listb

l1 = [10, 20, 30,  40, 300, 400, 999]
l2 = [10, 30,  100, 200, 300, 400]
l3 = list_union(l1, l2)

print l3
```

합집합이므로 lista와 listb를 합친 후 교집합을 제거해야 한다. 단순히 lista와 listb를 합친 후 중복되는 부분을 찾아 그 합친 리스트에서 제거하는 방법도 있지만, 이 방식은 낭비되는 코드가 있기 때문에 사용하지 않았다. 본인은 lista와 listb 중 짧은 길이를 가진 리스트의 원소를 탐색하여, 중복되는 부분이 있으면 버리고 그렇지 않으면 나머지 리스트에 추가하는(append) 방식을 사용하였다. 추가하는 작업이 끝난 후에는 오름차순으로 정렬(sort)하였다. 

## 5. 경로에 해당하는 문자열 1개를 입력 받아 그 안에 디렉토리 경로명과 파일명을 분리하여 리스트로 반환하는 함수 div_path(s)를 작성하시오.

```python
def div_path(s):
    n = len(s) # s의 길이 반환
    i=0; j=0; k=0 # 반복문 인자 
    pname = '\0' # 경로명
    fname = '\0' # 파일명
    
    while i<n:
        if s[n-i-1] == '/': #주어진 문자열의 가장 마지막 원소부터 첫 원소까지 차례대로 검색하며 '\' 문자를 찾습니다.
            break; #첫 번째 '\' 문자를 찾은 후 검색을 종료합니다.
            
        i+=1
    
    for j in range(n-i-1): #문자열 맨 처음부터 가장 마지막 '\' 문자 앞까지가 경로명입니다.
        pname += s[j]
    
    for k in range(n-i, n): #가장 마지막 '\' 문자의 뒷부분은 파일명입니다.
        fname += s[k]
    
    print "경로명: " + pname + "\n" + "파일명: " + fname
    
    return #return값이 없어도 써놔야 하는지 몰라서 그냥 썼습니다.

div_path('/usr/local/bin/python')
print
div_path('/home/chulsoo/test.txt')
    
    
```

가장 마지막 '\' 문자를 기준으로 그 앞은 경로명, 뒤는 파일명이라는 방식을 사용하여 문제를 해결하였습니다. 가장 마지막 '\' 문자를 찾기위해 문자열의 길이를 구하고 그 문자열의 가장 마지막 원소부터 첫 번째 원소 방향까지 반복문을 사용하여 차례대로 검색하는 방식을 사용하였습니다.

## 6. 웹 URL로 지정된 웹페이지를 문자열로 가져와 1) 모든 HTML 태그 및 CSS와 Javascript를 제외한 순수 텍스트만을 출력하고, 2) 그 순수 텍스트 안에 존재하는 단어가 총 몇개인지 출력하는 프로그램을 작성하시오.

```python
from UserString import MutableString #수정가능한(mutable) 문자열 변수를 생성하기 위함
import urllib

temp= urllib.urlopen('http://www.python.org/').read() #url 문자열을 임시 저장
s = MutableString(temp)
i=0 #반복문 인자
n = len(s) # s문자열의 길이
word_num=0 #단어 개수

html_write=False #html 태그 사이의 순수한 텍스트를 저장할 때 사용하는 조건 변수
script_tag=0 #javascript 태그가 사용된 수를 나타내는 변수
css_tag=0 #CSS 태그가 사용된 수를 나타내는 변수 / CSS 태그로는 span, li, li class 등이 있다. 

ptext="<pure text>\n" #true text 순수 택스트를 저장하는 문자열
ptext_num=0 #순수 텍스트의 개수

while i<n:
    if s[i] == '<':
        html_write=False
        
        if (s[i+1]=='s' and s[i+2]=='c' and s[i+3]=='r' and s[i+4]=='i' and s[i+5]=='p' and s[i+6]=='t')\
        or (s[i+1]=='S' and s[i+2]=='C' and s[i+3]=='R' and s[i+4]=='I' and s[i+5]=='P' and s[i+6]=='T'):
            script_tag+=1 #<script~> 혹은 <SCRIPT~>로 쓰였다면 스크립트 태그의 사용횟수를 1만큼 늘린다.
        elif (s[i+1]=='s' and s[i+2]=='t' and s[i+3]=='y' and s[i+4]=='l' and s[i+5]=='e')\
            or (s[i+1]=='S' and s[i+2]=='T' and s[i+3]=='Y' and s[i+4]=='L' and s[i+5]=='E'):
            css_tag+=1 #<style~> 혹은 <STYLE~>로 쓰였다면 CSS 태그의 사용횟수를 1만큼 늘린다.
        elif s[i+1]=='/':
            if (s[i+2]=='s' and s[i+3]=='c' and s[i+4]=='r' and s[i+5]=='i' and s[i+6]=='p' and s[i+7]=='t' and s[i+8]=='>')\
                or (s[i+2]=='S' and s[i+3]=='C' and s[i+4]=='R' and s[i+5]=='I' and s[i+6]=='P' and s[i+7]=='T' and s[i+8]=='>'):
                script_tag-=1 ##</script> 혹은 </SCRIPT>로 쓰였다면 스크립트 태그의 사용횟수를 1만큼 줄인다.
            elif (s[i+2]=='s' and s[i+3]=='t' and s[i+4]=='y' and s[i+5]=='l' and s[i+6]=='e')\
                or (s[i+2]=='S' and s[i+3]=='T' and s[i+4]=='Y' and s[i+5]=='L' and s[i+6]=='E'):
                css_tag+=1 #<style~> 혹은 <STYLE~>로 쓰였다면 CSS 태그의 사용횟수를 1만큼 줄인다.  
    elif s[i]=='>':
        if script_tag==0 and css_tag==0: #javascript 태그와 CSS 태그가 아닌 html태그 사이에 잇는 경우이다.
            html_write=True #html 태그 사이의 순수 텍스트를 추출하기로 한다.
            i+=1
            continue
    
    if html_write==True: #순수 텍스트 추출이 허가된 경우
        if s[i-1]=='>': #html 태그 사이의 첫 문자의 경우이다.
            if s[i] == ' ' or s[i]=='\n' or s[i]=='\t' or s[i-1]=='\b': 
                #html 태그 사이라도 의미없는 공백문자가 들어간 경우가 많다. 이경우는 추출하지 않는다.
                html_write=False
            else:
                ptext+='\n'
                ptext+=s[i] #순수 텍스트를 ptext 변수에 기록한다.
                word_num+=1 #순수 텍스트 출력과 동시에 단어 개수를 카운트한다.
        elif s[i-1]=='\n' or s[i-1]=='\t' or s[i-1]=='\b':
            i+=1
            continue
        elif s[i]==' ': #html 태그 사이에서 문자를 추출하는 도중 공백문자가 추출된 경우이다.
            if s[i-1]!=' ' and s[i-1]!='\n' and s[i-1]!='\t' and s[i-1]!='\b': #연속된 공백문자가 아닌 경우만 추출한다.
                ptext+='\n' #이 문제는 공백문자를 기준으로 단어를 나누기 때문에 공백문자가 나온 경우 개행한다.
                word_num+=1 #개행과 동시에 단어 개수를 카운트한다.
        elif s[i]!='\n' : #개행문자는 추출하지 않는다.
            ptext+=s[i] #순수 텍스트를 ptext 변수에 기록한다.
        
    i+=1    

print ptext
print "\n단어 개수: ", word_num, "개"
```

url 문자열을 분석한 결과 css의 span코드, li코드, li class코드와 script코드를 모두 삭제하고 html태그 사이의 텍스트만 추출하면 된다는 것을 이용하였다. 텍스트를 원본 문자열에서 추출하여 순수 텍스트를 담는 별개의 문자열 변수에 저장하는 방식을 사용하였다. 자세한 내용은 코드에 주석으로 적어 두었다. 단어를 나열하는데 중간중간 공백문자가 많이 보인다. 이것을 고치려고 많은 시간을 들여 노력했으나 잘 고쳐지지 않았다.
